# Data Cleaning

## Check for Missing Values

```python
missing = (df.isnull().sum()/len(df))*100

missing.sort_values(ascending=False)
```

### Replace missing values in the specific column

```python

df['Reasons for Churn'] = df['Reasons for Churn'].fillna('Active Customer')
```

## Check for Duplicate Rows



```python
df.drop_duplicates(inplace=True)
```

## Change Datatypes

```python
df["Date of Purchase"] = pd.to_datetime(
    df["Date of Purchase"],
    format="%b-%y"
)
```

## Remove spaces in text columns.

```python
text_columns = [
    "State",
    "MTN Device",
    "Gender",
    "Customer Review",
    "Subscription Plan",
    "Customer Churn Status",
    "Reasons for Churn"
]

for col in text_columns:
    df[col] = df[col].str.strip()
```

## Create Age Groups

```python
bins = [0, 24, 40, 60, 81]
labels = ["Gen Z (16-24)", "Millennials (25-40)", "Gen X (41-60)", "Seniors (61+)"]

df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)
```

## Create Customer Segment

```python
conditions = [
    (df["Customer Churn Status"] == "Yes"),  # Already left
    (df["Customer Tenure in months"] >= 24) & (df["Satisfaction Rate"] >= 4),  # Long-term & happy
    (df["Total Revenue"] >= df["Total Revenue"].median())  # High spenders
]

choices = [
    "Lost Customer",
    "Loyal Tier",
    "High Value"
]

df["Customer Segment"] = np.select(conditions, choices, default="Regular")
```
